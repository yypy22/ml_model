```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster, preprocessing
from sklearn import datasets

X, z = datasets.make_moons(n_samples=200, noise=0.08, random_state =0)
sc = preprocessing.StandardScaler()
X_norm=sc.fit_transform(X)

x=X_norm[:,0]
y=X_norm[:,1]
plt.figure(figsize=(10,3))
plt.scatter(x,y,c=z)
plt.show()
```

```python
spc=cluster.SpectralClustering(n_clusters=2, eigen_solver='arpack',
                                          affinity="nearest_neighbors")
# spc=cluster.SpectralClustering(n_clusters=2, affinity="rbf")
z_spc=spc.fit(X_norm)

plt.figure(figsize=(10,3))
plt.scatter(x,y,c=z_spc.labels_)
plt.show()
```
